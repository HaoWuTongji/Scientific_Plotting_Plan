```python
## hyper-parameter tunning ##

import numpy as np
import pandas as pd
import seaborn as sns  #  seaborn
import matplotlib.pyplot as plt

sns.set(style="white")

# 
alpha_values = [0.01, 0.1, 1]
beta_values = [0.95, 0.995, 0.9995]
lambda_values = [r'$0.1^2$', 0.1, r'$\sqrt{0.1}$']

rdi_alpha = np.array([42.19717656, 39.12294703, 52.61749867])  # 

rdi_beta = np.array([40.70317162, 40.26219711, 52.97225352])    # 

rdi_lambda = np.array([46.39363008, 40.01343304, 47.53055913])  # 


# 
plt.figure(figsize=(15, 5))

# 
alpha_positions = np.arange(len(alpha_values))
beta_positions = np.arange(len(beta_values))
lambda_positions = np.arange(len(lambda_values))

#   
plt.subplot(1, 3, 1)
sns.lineplot(x=alpha_positions, y=rdi_alpha, marker='o')
plt.xlabel(r'$\alpha$', fontsize=18)
plt.ylabel(r'$\overline{RDI}$', fontsize=18)
plt.xticks(alpha_positions, alpha_values, fontsize=16)  # 
plt.yticks(fontsize=14)  # y
plt.ylim(35, 55)  # y

#   
plt.subplot(1, 3, 2)
sns.lineplot(x=beta_positions, y=rdi_beta, marker='o', color='orange')
plt.xlabel(r'$\beta$', fontsize=18)
plt.ylabel('')  # y
plt.xticks(beta_positions, beta_values, fontsize=16)  # 
plt.yticks(fontsize=14)  # y
plt.ylim(35, 55)

#   
plt.subplot(1, 3, 3)
sns.lineplot(x=lambda_positions, y=rdi_lambda, marker='o', color='green')
plt.xlabel(r'$\lambda$', fontsize=18)
plt.ylabel('')  # y
plt.xticks(lambda_positions, lambda_values, fontsize=16)  # 
plt.yticks(fontsize=14)  # y
plt.ylim(35, 55)

# 
plt.tight_layout()

# 
save_path = 'D:/tuning1.svg'  # SVG

# SVG
plt.savefig(save_path, format='svg')

plt.show()
```
