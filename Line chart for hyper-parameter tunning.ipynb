```python
## 算法超参数调优（hyper-parameter tunning）的结果展示 ##

import numpy as np
import pandas as pd
import seaborn as sns  # 确保安装了 seaborn
import matplotlib.pyplot as plt

sns.set(style="white")

# 创建数据
alpha_values = [0.01, 0.1, 1]
beta_values = [0.95, 0.995, 0.9995]
lambda_values = [r'$0.1^2$', 0.1, r'$\sqrt{0.1}$']

rdi_alpha = np.array([42.19717656, 39.12294703, 52.61749867])  # α的示例数据

rdi_beta = np.array([40.70317162, 40.26219711, 52.97225352])    # β的示例数据

rdi_lambda = np.array([46.39363008, 40.01343304, 47.53055913])  # λ的示例数据


# 创建图形
plt.figure(figsize=(15, 5))

# 设置等间距的横坐标位置
alpha_positions = np.arange(len(alpha_values))
beta_positions = np.arange(len(beta_values))
lambda_positions = np.arange(len(lambda_values))

# 绘制 α 的折线图
plt.subplot(1, 3, 1)
sns.lineplot(x=alpha_positions, y=rdi_alpha, marker='o')
plt.xlabel(r'$\alpha$', fontsize=18)
plt.ylabel(r'$\overline{RDI}$', fontsize=18)
plt.xticks(alpha_positions, alpha_values, fontsize=16)  # 设置横轴刻度为指定的α值
plt.yticks(fontsize=14)  # 设置所有子图的y轴刻度字体大小
plt.ylim(35, 55)  # 设置y轴范围

# 绘制 β 的折线图
plt.subplot(1, 3, 2)
sns.lineplot(x=beta_positions, y=rdi_beta, marker='o', color='orange')
plt.xlabel(r'$\beta$', fontsize=18)
plt.ylabel('')  # 隐藏y轴标签
plt.xticks(beta_positions, beta_values, fontsize=16)  # 设置横轴刻度为指定的β值
plt.yticks(fontsize=14)  # 设置所有子图的y轴刻度字体大小
plt.ylim(35, 55)

# 绘制 λ 的折线图
plt.subplot(1, 3, 3)
sns.lineplot(x=lambda_positions, y=rdi_lambda, marker='o', color='green')
plt.xlabel(r'$\lambda$', fontsize=18)
plt.ylabel('')  # 隐藏y轴标签
plt.xticks(lambda_positions, lambda_values, fontsize=16)  # 设置横轴刻度为指定的λ值
plt.yticks(fontsize=14)  # 设置所有子图的y轴刻度字体大小
plt.ylim(35, 55)

# 调整布局
plt.tight_layout()

# 指定保存路径
save_path = 'D:/tuning1.svg'  # SVG文件保存路径

# 保存图像为SVG格式
plt.savefig(save_path, format='svg')

plt.show()
```
