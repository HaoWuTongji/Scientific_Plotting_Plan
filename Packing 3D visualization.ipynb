```python
## 装箱子结果的三维可视化 ##

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection  ## 确保安装了 mpl_toolkits

# 画出一个三维立方体
def draw_cube(ax, origin, size, color, alpha):
    # 立方体的8个顶点
    x, y, z = origin
    dx, dy, dz = size
    vertices = np.array([
        [x, y, z],
        [x+dx, y, z],
        [x+dx, y+dy, z],
        [x, y+dy, z],
        [x, y, z+dz],
        [x+dx, y, z+dz],
        [x+dx, y+dy, z+dz],
        [x, y+dy, z+dz]
    ])
    
    # 立方体的六个面（四个顶点确定一个面）
    faces = [
        [vertices[0], vertices[1], vertices[2], vertices[3]],  # 底面
        [vertices[4], vertices[5], vertices[6], vertices[7]],  # 顶面
        [vertices[0], vertices[1], vertices[5], vertices[4]],  # 前面
        [vertices[2], vertices[3], vertices[7], vertices[6]],  # 后面
        [vertices[1], vertices[2], vertices[6], vertices[5]],  # 右面
        [vertices[0], vertices[3], vertices[7], vertices[4]]   # 左面
    ]
    
    # 绘制每个面，设置颜色和边框
    ax.add_collection3d(Poly3DCollection(faces, facecolors=color, linewidths=0.5, edgecolors='k', alpha=alpha))
```

```python
# 初始化三维空间
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 三维空间的大小
space_size = [280, 280, 365]

# 绘制空间的边界框
ax.set_xlim([0, space_size[0]])
ax.set_ylim([0, space_size[1]])
ax.set_zlim([0, space_size[2]])

# 若干个立方体的左下角坐标和尺寸
cubes = [
    {'origin': [0.0, 69.0, 0], 'size': [41.9, 41.9, 7.0], 'color': '#a3f8b2'},# part 9 #a3f8b2
    {'origin': [41.9, 69.0, 0], 'size': [34.9, 34.3, 38.1], 'color': '#a0b0b2'}, # part 19 #a0b0b2
    {'origin': [0, 0, 0], 'size': [169.0, 69.0, 36.6], 'color': '#93cbf2'},# part 3 #93cbf2
    {'origin': [169, 0, 0], 'size': [85.3, 75.9, 46.0], 'color': '#ed91b5'}, # part 14 #ed91b5
    {'origin': [254.3, 0, 0], 'size': [14.9, 26.7, 22.4], 'color': '#97cff7'}, # part 2 #97cff7
]

# 绘制所有立方体
for cube in cubes:
    draw_cube(ax, cube['origin'], cube['size'], cube['color'], alpha=0.5)

# 绘制全透明的三维空间立方体（大尺寸的透明立方体）
draw_cube(ax, origin=[0, 0, 0], size=space_size, color='grey', alpha=0)

# 设置视角，调整透视角度
ax.view_init(elev=30, azim=-50)  # 调整观察角度，elev是仰角，azim是方位角

# 保存为SVG格式
# plt.savefig(r'D:\M2B1_250225.svg', format='svg')

# 显示图形
plt.show()
```
