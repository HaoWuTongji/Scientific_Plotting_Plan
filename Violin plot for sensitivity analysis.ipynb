```python
## 小提琴图，用于展示响应对参数的敏感度 #

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")

# 读取CSV数据
file_path = 'D:/data.csv'  # 请确认文件路径
data = pd.read_csv(file_path, header=None)

# 添加列名
data.columns = [r'$n_i$',r'$n_j$',r'$TF$', r'$RDD$', 'response1', 'response2']

# 定义绘图函数
def plot_combined_violin(data, response_column, factor):
    plt.figure(figsize=(10, 6))
    sns.violinplot(x=factor, y=response_column, data=data)
    plt.title(f'{response_column} vs {factor}')
    plt.xlabel(factor)
    plt.ylabel(response_column)
    plt.grid(True)
    plt.show()

# 绘制响应1的小提琴图
factors = [r'$n_i$',r'$n_j$',r'$TF$', r'$RDD$']
for factor in factors:
    plot_combined_violin(data, 'response1', factor)

# 绘制响应2的小提琴图
for factor in factors:
    plot_combined_violin(data, 'response2', factor)
```

```python
## 增加了截断区域 ##

sns.set(style="darkgrid")

# 读取CSV数据
file_path = 'D:/data1.csv'  # 请确认文件路径
data = pd.read_csv(file_path, header=None)

# 添加列名
data.columns = [r'$n_i$',r'$n_j$',r'$TF$', r'$RDD$', 'response1']

# 定义绘图函数
def plot_combined_violin(data, response_column):
    factors = [r'$n_i$',r'$n_j$',r'$TF$', r'$RDD$']
    
    fig, axes = plt.subplots(1, len(factors), figsize=(20, 6), sharey=True)

    for ax, factor in zip(axes, factors):
        sns.violinplot(x=factor, y=response_column, data=data, ax=ax, bw=0.3, cut=1)  ## 控制截断区域和图形胖瘦 ##
        ax.set_xlabel(factor)
        ax.set_ylabel('')  # 设置y轴标签为空
        ax.grid(True)

        # 设置y轴为百分数
        ax.yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1.0))

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    
    # 保存为SVG格式
#     plt.savefig(r'D:\violin_plot241106.svg', format='svg')
    plt.show()

# 绘制响应1的小提琴图
plot_combined_violin(data, 'response1')
```
